```python
a = np.linspace(-100, 100, 20)
a_sq = a ** 2
```

```python
plt.figure(figsize=(6, 4))
plt.plot(a, a_sq, '.-', color='#800020')
plt.plot([-100, 100], [2000, 8000], 'y*-')
plt.title('Parabola', size=25, color='#800020')
plt.xticks(rotation=45, size=12)   #yticks
# plt.ylim(0, 400)
plt.legend(['SQ', 'LN'])
plt.grid()
plt.show()
```
